```python
import urllib.request
import xlrd
import csv
import pandas as pd

#urlretrieve를 이용하여 파일에 직접 저장
url = "https://raw.githubusercontent.com/jonghunwoo/public/master/block_transfer_record.xlsx"
print(url)
filename = "block_transfer_list.xlsx"
urllib.request.urlretrieve(url, filename)

#uropen을 이용하여 메모리에 저장후 파일에 저장
"""
url = "http://snowdeer.github.io/public/img/hello_page.jpg"
filename = "snowdeer.jpg"

image = urllib.request.urlopen(url).read()

with open(filename, mode="wb") as f:
    f.write(image)
"""    
```

```python
#xls를 csv로 변환하는 함스
def csv_from_excel():

    wb = xlrd.open_workbook('block_transfer_list.xlsx')
    sh = wb.sheet_by_name('record')
    csv_file = open('block_transfer_list.csv', 'w')
    wr = csv.writer(csv_file, quoting=csv.QUOTE_ALL)

    for rownum in range(sh.nrows):
        wr.writerow(sh.row_values(rownum))

    csv_file.close()
    
csv_from_excel()

data = pd.read_csv('block_transfer_list.csv')
```
